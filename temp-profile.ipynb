```python
import numpy as np
import matplotlib.pyplot as plt

# temperature profile
T0 = 300  # K
p0 = 100000  # Pa
g = 9.81
Rd = 287
lev = np.arange(0, 12400, step=400)


def Temp(gamma):
    T = np.zeros(len(lev))
    for i in range(len(lev)):
        p = p0 * (1 - (gamma * lev[i]) / T0) ** (g / (Rd * gamma))
        T[i] = T0 * (p / p0) ** ((Rd * gamma) / g)

    return T


fig, ax = plt.subplots()
ax.plot(Temp(0.0065), lev)
ax.plot(Temp(0.0090), lev)
```
